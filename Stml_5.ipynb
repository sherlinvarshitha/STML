```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
salary=pd.read_csv('/content/Salary_Data.csv')
salary.head()
```

```python
features=salary.drop('Salary',axis=1)
features.head()
```

```python
x_normalized = features['YearsExperience']-features['YearsExperience'].min()/features['YearsExperience'].max()-features['YearsExperience'].min()
# fix the typo in the variable name
x_normalized.head()

target_var = salary['Salary']
plt.scatter(features['YearsExperience'], target_var)
plt.xlabel('Years of Experience')
plt.ylabel('Salary Variable')
plt.title('Relationship between Years of Experience and salary')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
diabeits = pd.read_csv('/content/diabetes_data_upload.csv')
diabeits.head()
```

```python
target_var=diabeits['class']
target_var.head()
```

```python
features=diabeits.drop('class',axis=1)
features.head()
```

```python
for col in features.columns:
  plt.figure(figsize=(10,6))
  plt.scatter(features[col],target_var)
  plt.xlabel(col)
  plt.ylabel('Target')
  plt.title(f'Relationship between {col} and Target')
  plt.show()
```
