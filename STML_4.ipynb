Q1

```python
import pandas as pd
df=pd.read_csv('/content/train.csv')
df
```

```python
y=df['price_range']
y
```

```python
x=df.drop('price_range',axis=1)
x
```

```python
upd=x-x.min()/x.max()-x.min()
upd
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(upd,y,test_size=0.7,random_state=30)
print(x_train)
print(x_test)
print(y_train)
print(y_test)
```
