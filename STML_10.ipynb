Q1

```python
 import pandas as pd
 from sklearn.model_selection import train_test_split
 from sklearn.neighbors import KNeighborsClassifier
 from sklearn.svm import SVC
 from sklearn.metrics import accuracy_score
 import zipfile

data = pd.read_csv('/content/breast_cancer_survival.csv')
data.head()
```

```python
 data_encoded = pd.get_dummies(data, drop_first=True)
 X = data_encoded.drop(columns=['Age'])
y = data_encoded['Age']
 test_sizes = [0.2, 0.25, 0.3, 0.35]
 knn_accuracies = {}
 for size in test_sizes:

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=size, random_state=42)

    knn = KNeighborsClassifier(n_neighbors=5)
    knn.fit(X_train, y_train)


    y_pred = knn.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    knn_accuracies[size] = accuracy
    print(f"KNN - Test Size: {size}, Accuracy: {accuracy:.4f}")
 print("KNN Accuracies with different test sizes:", knn_accuracies)
```

Q2

```python
svm_accuracies = {}
for size in test_sizes:
    # Indented block for the for loop
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=size, random_state=42)
    svm = SVC(kernel='linear')
    svm.fit(X_train, y_train)
    y_pred = svm.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    svm_accuracies[size] = accuracy
    print(f"SVM - Test Size: {size}, Accuracy: {accuracy:.4f}")
print("SVM Accuracies with different test sizes:", svm_accuracies)
```
